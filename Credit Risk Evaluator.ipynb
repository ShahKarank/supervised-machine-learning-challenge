```python
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.model_selection import train_test_split
```

```python
# Import the data
data = pd.read_csv('Resources/lending_data.csv')
data
```

<b> I believe Random Forest Classifier will perform better due to its tendency to not assume and linear relationship and feature selection which shall provide it an edge over Logistic Regression, however due to the lack of noise variables. Logistic Regression might be a strong contender. </b>

```python
X = data.drop('loan_status', axis=1)
y = data['loan_status'].values
X_train, X_test, y_train, y_test = train_test_split(X,y)
X_train
```

```python
# Train a Logistic Regression model print the model score
from sklearn.linear_model import LogisticRegression
cf = LogisticRegression(max_iter=10000,random_state=42)
cf.fit(X_train, y_train)
print(f"Train Score: {cf.score(X_train, y_train)}")
print(f"Test Score: {cf.score(X_test, y_test)}")
```

```python
# Train a Random Forest Classifier model and print the model score
from sklearn.ensemble import RandomForestClassifier
crf = RandomForestClassifier(random_state=42, n_estimators=500)
crf.fit(X_train, y_train)
print(f"Train Score: {crf.score(X_train, y_train)}")
print(f"Test Score: {crf.score(X_test, y_test)}")
```

<b> Turns out I stand corrected. Contrary to my Prediction, Logistic regression outperformed Random Forest Classifier by a razor thin Margin. Scores stand at (LR = 0.9920 and RFC = 0.9917). In such an instance both models seem equally viable with minimal differences. </b>
